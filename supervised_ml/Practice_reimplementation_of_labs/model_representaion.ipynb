```python
import numpy as np
import matplotlib.pyplot as plt
```

Training Data

```python
#Creating training data
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
print(f'x_train: {x_train}')
print(f'y_train: {y_train}')
```

```python
print(f'Shape of x_train: {x_train.shape}')
print(f'Shape of y_train: {y_train.shape}')
```

```python
#Calculating number of training examples

#1st method
m = x_train.shape[0]
print(f'Number of training examples: {m}')

#2nd method
m = len(x_train)
print(f'Number of training examples: {m}')
```

Plotting the training data

```python
plt.scatter(x_train, y_train, marker='x', c='r')
plt.title('Housing Prices')
plt.xlabel('House size in 1000 sqft')
plt.ylabel('House price in 1000 dollars')
```

Weights and Biases

```python
# Declaring weigts and biases
w, b = 100, 100
print(f'weight (w) = {w}, bias (b) = {b}')
```

Model Function

```python
# Samples values from training set
i = 0
x_i, y_i = x_train[i], y_train[i]
print(f'x^{i} = {x_i}, y^{i} = {y_i}')
f_wb = w * x_i + b
print(f'Model output for sample training example: {f_wb}')
```

```python
def calculate_model_output(x, w, b):
    m = x.shape[0]
    f_wb = np.zeros(m)

    for i in range(m):
        f_wb[i] = w * x[i] + b
    
    return f_wb
```

```python
def plot_data(w, b):
    #Calculating Prediction   
    f_wb = calculate_model_output(x_train, w, b)
    print(f'f_wb = {f_wb}')

    #Plotting Prediction
    plt.plot(x_train, f_wb, c='b', label='Prediction')

    #Plotting data point
    plt.scatter(x_train, y_train, marker='x', c='r', label='Actual Values')

    plt.title('Housing Prices Prediction')
    plt.xlabel('Size of House in 1000 sqft')
    plt.ylabel('Price of House in 1000 dollars')
    plt.legend()
    plt.show()
```

Trying different values of w and b to get a good model

```python
plot_data(w=100, b=100)
```

```python
plot_data(w=150, b=100)
```
